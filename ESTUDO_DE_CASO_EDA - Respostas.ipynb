### **QUESTÃO 2**

```python
# Mostrar os principais produtos agrupados por seção e categorias.
database_questao2_agrup = database_questao2.groupby(['SECAO_PRODUTO', 'CATEGORIA_PRODUTO']).agg(
        PRECO_UNITARIO_MEDIA=('PRECO_UNITARIO', 'mean'),
        QTD_UNIDADES_VENDIDAS_MES_SUM=('QTD_UNIDADES_VENDIDAS_MES', 'sum'),
        FATURAMENTO_BRUTO_SUM=('FATURAMENTO_BRUTO', 'sum'),
        FATURAMENTO_BRUTO_PERCENT_SUM=('FATURAMENTO_BRUTO%', 'sum')
).round(2).sort_values(by=["SECAO_PRODUTO", "FATURAMENTO_BRUTO_SUM"], ascending=[True, False])

# Mostrar os 5 maiores de cada seção pela ordenaçao do faturamento
database_questao2_agrup = database_questao2_agrup.groupby('SECAO_PRODUTO').head(5)
database_questao2_agrup

```

### **QUESTÃO 3**

#### Existe relação entre o preço unitário e a quantidade de unidades vendidas? Essa relação muda a depender da seção do produto?

```python
database_questao3 = database.copy()

# Selecionando as variáveis
vendas =  database_questao3['QTD_UNIDADES_VENDIDAS_MES']
preco = database_questao3['PRECO_UNITARIO']

# Montando os cortes, escolhendo os quartis

cortes = [
    0,
    0.01,
    vendas.quantile(0.25),
    vendas.quantile(0.50) ,
    vendas.quantile(0.75),
    vendas.max()]

labels_qtd = ['Nenhum', 'Baixa', 'Média', 'Alta', 'Muito Alta']

database_questao3['QTD_UNIDADES_VENDIDAS_MES_FAIXAS'] = pd.cut(
    vendas,
    labels=labels_qtd,
    bins=cortes,
    include_lowest=True)

cortes = [
    0,
    preco.quantile(0.25),
    preco.quantile(0.50),
    preco.quantile(0.75),
    preco.max()]

labels_preco = ['Muito Barato','Barato','Preço Médio','Caro']

database_questao3['PRECO_UNITARIO_FAIXAS']  = pd.cut(
    preco,
    bins=cortes,
    labels=labels_preco,
    include_lowest=True)

database_questao3['PRECO_UNITARIO_FAIXAS'].value_counts().sort_index()
```

```python
database_questao3
```

```python
# cruzamento entre as 2 variaveis, criando uma tabela de frequência
pd.crosstab(
    database_questao3['QTD_UNIDADES_VENDIDAS_MES_FAIXAS'],
    database_questao3['PRECO_UNITARIO_FAIXAS'],
    normalize = 'columns',
    margins=True,
    margins_name='Total').round(2) * 100
```

```python
#cruzamento entre as 2 variaveis, agrupando por secao
pd.crosstab(
    index=[database_questao3['SECAO_PRODUTO'], database_questao3['QTD_UNIDADES_VENDIDAS_MES_FAIXAS']],
    columns=database_questao3['PRECO_UNITARIO_FAIXAS'],
    normalize = 'columns',
    margins=True,
    margins_name='Total').round(2) * 100
```

```python
# Isolando a seção "Acessórios"
database_questao3_filter = database_questao3[database_questao3['SECAO_PRODUTO'] == 'Acessórios']
pd.crosstab(database_questao3_filter['QTD_UNIDADES_VENDIDAS_MES_FAIXAS'], database_questao3_filter['PRECO_UNITARIO_FAIXAS'], normalize = 'columns', margins=True, margins_name='Total').round(2) * 100
```

```python
# Isolando a seção "Informatica"
database_questao3_filter = database_questao3[database_questao3['SECAO_PRODUTO'] == 'Informática']
pd.crosstab(database_questao3_filter['QTD_UNIDADES_VENDIDAS_MES_FAIXAS'], database_questao3_filter['PRECO_UNITARIO_FAIXAS'], normalize = 'columns', margins=True, margins_name='Total').round(2) * 100
```

```python
# Isolando a seção "Papelaria"
database_questao3_filter = database_questao3[database_questao3['SECAO_PRODUTO'] == 'Papelaria']
pd.crosstab(database_questao3_filter['QTD_UNIDADES_VENDIDAS_MES_FAIXAS'], database_questao3_filter['PRECO_UNITARIO_FAIXAS'], normalize = 'columns', margins=True, margins_name='Total').round(2) * 100
```

Utilizando a segmentação por quartis para as variáveis PRECO_UNITARIO e QTD_UNIDADES_VENDIDAS_MES, foi possível observar — por meio de tabelas de frequência de dupla entrada — uma relação inversamente proporcional entre o preço e o volume de vendas.

Analisando a tabela geral de forma mais detalhada, nota-se que a faixa de preço classificada como "Muito Barato" está fortemente concentrada nas categorias de quantidade "Alta" e "Muito Alta", indicando que produtos com preços mais acessíveis tendem a apresentar alta rotatividade. Esse comportamento é valioso para o planejamento de estoque, pois sugere que esses produtos devem ter reposição constante.

Por outro lado, os produtos na faixa de preço "Caro" estão majoritariamente associados às faixas de quantidade "Baixa" e "Média", reforçando o comportamento inverso — produtos mais caros tendem a ter uma saída mais lenta.

Ao segmentar essa análise por seção, temos:

### **QUESTÃO 4**

#### Considerando os produtos com maior faturamento bruto, quantos deles foram responsáveis por 80% do faturamento bruto total da loja?

```python
database_questao4 = database.copy()
database_questao4["FATURAMENTO_BRUTO"] = database_questao4["PRECO_UNITARIO"] * database_questao4["QTD_UNIDADES_VENDIDAS_MES"]
database_questao4_ordenada = database_questao4.sort_values(by="FATURAMENTO_BRUTO", ascending=False)

# Calcular faturamento acumulado, arredondando por 3 casas
database_questao4_ordenada["FATURAMENTO_BRUTO_ACUMULADO"] = database_questao4_ordenada["FATURAMENTO_BRUTO"].cumsum()
faturamento_total = database_questao4_ordenada["FATURAMENTO_BRUTO"].sum()
database_questao4_ordenada["FATURAMENTO_PCT_ACUMULADO"] = (database_questao4_ordenada["FATURAMENTO_BRUTO_ACUMULADO"] / faturamento_total).round(3)

# Contar quantos produtos acumulam até 80% do faturamento
produtos_80_fat = database_questao4_ordenada[(database_questao4_ordenada["FATURAMENTO_PCT_ACUMULADO"] <= 0.80)]

qtd_faturamento80 = len(produtos_80_fat)
qtd_total = len(database_questao4)

# quantidade da base filtrada
print(f"A base filtrada possui {qtd_faturamento80} produtos")
print(f"A base total possui {qtd_total} produtos")

porcent_produtos = round((qtd_faturamento80 / qtd_total),3)  * 100

print(f"Isso representa {porcent_produtos}% dos produtos")
```

Considerando o faturamento bruto individual dos produtos (resultado do PRECO_UNITARIO multiplicado pela QTD_UNIDADES_VENDIDAS_MES) e, em seguida, calculando o faturamento acumulado e seu percentual em relação ao total da loja (com arredondamento para três casas decimais), foi identificado que 49 produtos são responsáveis por 80% do faturamento bruto total.

Dado que a base de dados completa possui 290 produtos, esses 49 produtos representam aproximadamente 16,9% do portfólio total. Este achado evidencia um claro comportamento de Efeito Pareto (também conhecido como regra 80/20), onde uma parcela relativamente pequena de produtos gera a maior parte da receita bruta da loja.

### **QUESTÃO 6**

#### Existem produtos com valores atípicos de quantidade de unidades vendidas? E de preço unitário? E de faturamento bruto?

```python
database_questao6 = database.copy()

# Selecionando as variáveis
unidades_vendidas = database_questao6['QTD_UNIDADES_VENDIDAS_MES']
preco_unitario = database_questao6['PRECO_UNITARIO']
faturamento_bruto = database_questao6['PRECO_UNITARIO'] * database_questao6['QTD_UNIDADES_VENDIDAS_MES']

# Criando método para imprimir os dados comum para análise das variáveis
def detalhes_variavel(var):
    print("O mínimo é", var.min())
    print("O percentil 1 é", var.quantile(0.01).round(1))
    print("O máximo é", var.max())
    print("O percentil 99 é", var.quantile(0.99).round(1))
    print("A amplitude é", var.max() - var.min())
    print("A amplitude percentílica é", (var.quantile(0.99) - var.quantile(0.01)).round(1))
    print("A média é", var.mean())
    print("A mediana é", var.median())

    limite_inferior = var.quantile(0.25) - 1.5 * (var.quantile(0.75) - var.quantile(0.25))
    limite_superior = var.quantile(0.75) + 1.5 * (var.quantile(0.75) - var.quantile(0.25))

    print(f"Limite Inferior: {limite_inferior}")
    print(f"Limite Superior: {limite_superior}")
    print(f"Regra Tradicional do Intervalo Interquartil Inferior: {var[var < limite_inferior].count()}")
    print(f"Regra Tradicional do Intervalo Interquartil Superior: {var[var > limite_superior].count()}")
```

```python
# Observar 20 menores da variavel QTD_UNIDADES_VENDIDAS_MES
unidades_vendidas.nsmallest(20)
```

```python
# Observar 20 maiores da variavel QTD_UNIDADES_VENDIDAS_MES
unidades_vendidas.nlargest(20)
```

```python
# imprimir os detalhes da variável
detalhes_variavel(unidades_vendidas)
```

```python
# Observar 20 menores da variavel PRECO_UNITARIO
preco_unitario.nsmallest(20)
```

```python
# Observar 20 maiores da variavel PRECO_UNITARIO
preco_unitario.nlargest(20)
```

```python
# Observar 20 maiores da variavel faturamento_bruto
faturamento_bruto.nlargest(20)
```

```python
# imprimir os detalhes da variável
detalhes_variavel(faturamento_bruto)
```

### **QUESTÃO 7**

#### Qual seção de produtos tem maior variabilidade de preços unitários?

```python
database_questao7 = database.copy()

# Selecionando variáveis
preco = database_questao7['PRECO_UNITARIO']
secao = database_questao7['SECAO_PRODUTO']

# Criação do dataframe com o describe do preço por seção
database_questao7_agrup = pd.DataFrame(preco.groupby(secao).describe())

# Criação das colunas com os cálculos para comparação
database_questao7_agrup['AMPLITUDE'] = database_questao7_agrup['max'] - database_questao7_agrup['min']
p01 = preco.groupby(secao).quantile(0.01)
p99 = preco.groupby(secao).quantile(0.99)
database_questao7_agrup['AMPLITUDE_PERCENTILICA'] = (p99 - p01).round(1)
desvio_abs_medio = preco.groupby(secao).apply(lambda x: np.mean(np.abs(x - x.mean())))
database_questao7_agrup['DESVIO_ABSOLUTO_MEDIO'] = desvio_abs_medio
desvio_abs_mediano = preco.groupby(secao).apply(lambda x: np.median(np.abs(x - np.median(x))))
database_questao7_agrup['DESVIO_ABSOLUTO_MEDIANO'] = desvio_abs_mediano
database_questao7_agrup['COEFICIENTE_VARIACAO'] = (database_questao7_agrup['std'] / database_questao7_agrup['mean']).round(3)  * 100
database_questao7_agrup['COEFICIENTE_VARIACAO_MEDIANO'] = (database_questao7_agrup['std'] / database_questao7_agrup['50%']).round(3)  * 100

# Exibir base
database_questao7_agrup
```

A análise da variabilidade dos preços unitários por seção, utilizando o Coeficiente de Variação (CV), revela que a seção Informática possui a maior variabilidade.

Coeficiente de Variação (CV) baseado na Média:<br>
Informática: 150,8%<br>
Papelaria: 147,6%<br>
Acessórios: 75,2%<br>
Inicialmente, com base no CV padrão (desvio padrão dividido pela média), as seções de Informática e Papelaria apresentam variabilidades próximas e elevadas, enquanto Acessórios mostra uma distribuição de preços mais homogênea.

No entanto, ao aprofundar a análise com o Coeficiente de Variação Mediano (desvio padrão dividido pela mediana, capturando melhor a assimetria e o impacto de valores extremos), a diferença se torna mais expressiva:

O CV Mediano de Informática aumenta drasticamente de 150,8% para 662,6%.<br>
O CV Mediano de Papelaria sobe de 147,6% para 271,2%.<br>
Essa discrepância acentuada entre o CV tradicional e o CV Mediano, particularmente na seção de Informática, indica uma maior assimetria e um impacto mais forte de valores extremos (outliers) nos preços unitários. Isso significa que, embora a Papelaria também seja heterogênea, a seção de Informática apresenta produtos com preços que se desviam muito mais da "normalidade" do seu centro (mediana), o que reforça sua posição como a seção com a maior variabilidade de preços unitários.
